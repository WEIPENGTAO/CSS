```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from file
data = np.loadtxt("performance_data.txt")

# Extract x and y values from data
x = data[:, 0]  # thread numbers
y = data[:, 1]  # elapsed time

# Plot performance curve
plt.plot(x, y, marker='o')

# Set axis labels and title
plt.xlabel("Thread Number")
plt.ylabel("Elapsed Time (s)")
plt.title("Performance Curve")

# Show plot
plt.show()
```

```python
data = []
with open('ematrix2.txt', 'r') as file:
    for line in file:
        data.append(line)
data = data[1:-1]
data
```

```python
import re

size = []
serial_time = []
parallel_time = []
for i in range(len(data) // 2):
    serial_list = data[i*2].split(",")
    parallel_list = data[i*2+1].split(",")
    for i in range(len(serial_list)):
        serial_list[i] = re.sub(r'[^\d.]', '', serial_list[i])
    for i in range(len(parallel_list)):
        parallel_list[i] = re.sub(r'[^\d.]', '', parallel_list[i])
    size.append(int(serial_list[0]))
    serial_time.append(float(serial_list[2]))
    parallel_time.append(float(parallel_list[2]))
```

```python
size
```

```python
serial_time
```

```python
parallel_time
```

```python
import matplotlib.pyplot as plt

# Calculate speedup
speedup = [serial_time[i] / parallel_time[i] for i in range(len(size))]

# Plot performance curves
plt.plot(size, serial_time, label="Serial")
plt.plot(size, parallel_time, label="Parallel (8 threads)")

# Add legend and title
plt.legend()
plt.title("Performance Comparison")

plt.savefig('Performance Comparison.png')

# Plot speedup curve
plt.figure()
plt.plot(size, speedup, label="Speedup (8 threads)")
plt.legend()
plt.title("Speedup Comparison")

plt.savefig('Speedup Comparison.png')

# Show the plots
plt.show()
```

```python
import pandas as pd

# Create a pandas data frame
df = pd.DataFrame({
    'Matrix Size': size,
    'Serial Time': serial_time,
    'Parallel Time': parallel_time
})

# Calculate speedup and add it to the data frame
df['Speedup'] = df['Serial Time'] / df['Parallel Time']

# Print the data frame
df
```
